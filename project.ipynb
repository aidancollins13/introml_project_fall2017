```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("MissingMigrants-Global-2017-12-11T17-38-26.csv")
df
```

```python
data = np.array(df)
causes = df['Cause of death'].unique()
print(causes)
x = np.column_stack((causes, np.zeros(len(causes))))
for row in data:
    for i in x:
        if row[12] == i[0]:
           i[1] += row[7]
```

```python


x = x[x[:,1].argsort()][::-1]
plt.bar(range(20), list(x[:20,1]))
plt.xticks(range(20), list(x[:20,0]),rotation='vertical')
plt.show()
```
